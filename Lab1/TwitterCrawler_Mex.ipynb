```python
#available libraries: TextBlob, VADER, Indico.io

from tweepy import Stream
from tweepy.streaming import StreamListener
from textblob import TextBlob
from pprint import pprint

class MyListener(StreamListener): #listen tweets
    
    def __init__(self, api = None): #initialize to 0 init function
        super(StreamListener, self).__init__()
        self.num_tweets = 0
        self.analized = {}  
        self.analized['tweets'] = []  
        
    def on_data(self, data): 
        try:
            with open(jsonFileName,'a') as f: #file name wherever you want #open the file and append (a) new tweet

                twitter_text = json.loads(data)['text'] #give the text of the tweet

                analysis = TextBlob(twitter_text)
                
                self.analized['tweets'].append({  
                    'id': json.loads(data)['id'],
                    'text': json.loads(data)['text'],
                    'analysis': analysis.sentiment
                })
                
                self.num_tweets += 1
                if self.num_tweets < 10000:
                    print("Saved new tweet number: %s" % self.num_tweets)
                    return True
                else:
                    #write the obtained data
                    json.dump(self.analized, f, indent=4)
                    print('done')
                    return False
        except BaseException as e:
            print("Error on_data: %s" % str(e))
        return True
    
    def on_error(self, status): #print the error
        print('Error :', status)
        return False
    

print('..starting the tweet gathering...')    
twitter_stream = Stream(auth, MyListener()) #initialize the stream

twitter_stream.filter(track=['border wall','crime'],languages=['en'],locations=[-116.48,14.05,-90.82,28.54])



```

```python
#move the data to final_result_path when you are happy with the results before moving on
import numpy

jsonPath = final_result_path + jsonFileName
polarities = []
subjectivities = []

with open(jsonPath) as json_file:  
    data = json.load(json_file)
    for tweet in data['tweets']:
        polarities.append(tweet['analysis'][0])
        subjectivities.append(tweet['analysis'][1])
        
  
polaritiesAverage = sum(polarities)/float(len(polarities))
polaritiesStandardDeviation = numpy.std(polarities)
subjectivitiesAverage = sum(subjectivities)/float(len(subjectivities))
subjectivitiesStandardDeviation = numpy.std(subjectivities)

print('polarities Average : %s' % polaritiesAverage)
print('polarities Standard Deviation : %s' % polaritiesStandardDeviation)
print('subjectivities Average : %s' % subjectivitiesAverage)
print('subjectivities Standard Deviation : %s' % subjectivitiesStandardDeviation)



print("")
print("----")
print("removing the zeroes ...")
print("----")
print("")


polaritiesNoZeros = []
subjectivitiesNoZeros = []

for i in range(len(polarities)):
    if(polarities[i]!=0.0 or subjectivities[i]!=0.0):
        polaritiesNoZeros.append(polarities[i])
        subjectivitiesNoZeros.append(subjectivities[i])
        
polaritiesNZAverage = sum(polaritiesNoZeros)/float(len(polaritiesNoZeros))
polaritiesNZStandardDeviation = numpy.std(polaritiesNoZeros)
subjectivitiesNZAverage = sum(subjectivitiesNoZeros)/float(len(subjectivitiesNoZeros))
subjectivitiesNZStandardDeviation = numpy.std(subjectivitiesNoZeros)

print("No zero values, length %s" %len(polaritiesNoZeros))
print('polarities Average : %s' % polaritiesNZAverage)
print('polarities Standard Deviation : %s' % polaritiesNZStandardDeviation)
print('subjectivities Average : %s' % subjectivitiesNZAverage)
print('subjectivities Standard Deviation : %s' % subjectivitiesNZStandardDeviation)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def sentiment_analyzer_score(sentence):
    score = analyzer.polarity_scores(sentence)
    return score['compound']
    
## we are interested in the compound value
jsonPath = final_result_path + jsonFileName
compounds = []

with open(jsonPath) as json_file:  
    data = json.load(json_file)
    for tweet in data['tweets']:
        compounds.append(sentiment_analyzer_score(tweet['text']))

        
compoundsAverage = sum(compounds)/float(len(compounds))
compoundsStandardDeviation = numpy.std(compounds)

print('compound Average with Vader is : %s' % compoundsAverage)
print('compound Standard Deviation with Vader is: %s' % compoundsStandardDeviation)

```
